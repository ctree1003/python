```python
from ast import literal_eval
```

```python
credits['cast'] = credits['cast'].apply(literal_eval)
```

```python
def get_name(x):
    for i in x:
        if i['name']
            return i['name']
    return []
```

```python
credits['cast'] = credits['cast'].apply(get_name)
```

```python
credits['crew'] = credits['crew'].apply(literal_eval)
```

```python
def get_Director(x):
    for i in x:
        if i['job']=='Director':
            return i['name']
    return []
```

```python
credits['crew'] = credits['crew'].apply(get_Director)
```
